

### **Introduction to Curve Fitting**

#### What is Curve Fitting?
Curve fitting involves finding a mathematical function that best describes the relationship between data points. Itâ€™s commonly used in astronomy to:
1. Analyze observational data.
2. Model physical phenomena.

#### Key Steps in Curve Fitting:
1. Define the model function.
2. Use a fitting algorithm to find the best parameters.
3. Evaluate the fit using residuals or error metrics.

#### Example 1: Linear Fit
Suppose we have data that roughly follows a straight line. We use the equation:
$$ y = mx + c $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 4.1, 6.2, 8.0, 10.1])

# Define the model function
def linear_model(x, m, c):
    return m * x + c

# Perform curve fitting
params, covariance = curve_fit(linear_model, x_data, y_data)

# Extract fitted parameters
m, c = params
print(f"Slope (m): {m}, Intercept (c): {c}")

# Plot the data and the fitted curve
plt.scatter(x_data, y_data, label="Data", color="red")
plt.plot(x_data, linear_model(x_data, m, c), label="Fitted Line", color="blue")
plt.title("Linear Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### **Advanced Curve Fitting**

#### Example 2: Fitting Non-Linear Data
Use the equation for a quadratic curve:
$$ y = ax^2 + bx + c $$

```python
# Define the model function
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Generate synthetic data
x_data = np.linspace(-10, 10, 50)
y_data = 3 * x_data**2 - 2 * x_data + 5 + np.random.normal(0, 10, size=x_data.shape)

# Perform curve fitting
params, covariance = curve_fit(quadratic_model, x_data, y_data)
a, b, c = params
print(f"Coefficients: a={a}, b={b}, c={c}")

# Plot the data and fitted curve
plt.scatter(x_data, y_data, label="Data", color="green", alpha=0.6)
plt.plot(x_data, quadratic_model(x_data, a, b, c), label="Fitted Curve", color="orange")
plt.title("Quadratic Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

#### Example 3: Polynomial Fit Using NumPy
Polynomial fits are useful for approximating data with polynomial functions. NumPy provides the `polyfit` function for this purpose.

```python
# Sample data
x_data = np.linspace(-5, 5, 50)
y_data = 2 * x_data**3 - 3 * x_data**2 + x_data + np.random.normal(0, 10, size=x_data.shape)

# Perform polynomial fitting (degree 3)
p_coeffs = np.polyfit(x_data, y_data, 3)
print(f"Polynomial Coefficients: {p_coeffs}")

# Generate the polynomial function
p_func = np.poly1d(p_coeffs)

# Plot the data and polynomial fit
plt.scatter(x_data, y_data, label="Data", color="blue", alpha=0.6)
plt.plot(x_data, p_func(x_data), label="Polynomial Fit", color="red")
plt.title("Polynomial Fit Using NumPy")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

#### Example 4: Polynomial Fit in Astronomy (Stellar Brightness)
Suppose we have observational data of a star's brightness over time:

```python
# Generate synthetic data for brightness over time
time = np.linspace(0, 10, 100)
brightness = -0.5 * time**2 + 3 * time + 10 + np.random.normal(0, 2, size=time.shape)

# Fit a quadratic polynomial
coeffs = np.polyfit(time, brightness, 2)
quad_func = np.poly1d(coeffs)

# Plot the data and the fit
plt.scatter(time, brightness, label="Observed Brightness", color="purple", alpha=0.6)
plt.plot(time, quad_func(time), label="Fitted Quadratic Curve", color="orange")
plt.title("Stellar Brightness Over Time")
plt.xlabel("Time (days)")
plt.ylabel("Brightness (arbitrary units)")
plt.legend()
plt.grid(True)
plt.show()
```

#### Example 5: Fitting Astronomical Data (Gaussian Fit)
Gaussian functions are commonly used to model spectral lines or light profiles:
$$ y = A e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $$

```python
# Define the Gaussian function
def gaussian(x, A, mu, sigma):
    return A * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Generate synthetic data
x_data = np.linspace(-5, 5, 100)
y_data = gaussian(x_data, 10, 0, 1) + np.random.normal(0, 0.5, size=x_data.shape)

# Perform curve fitting
params, covariance = curve_fit(gaussian, x_data, y_data, p0=[10, 0, 1])
A, mu, sigma = params
print(f"Fitted Parameters: A={A}, mu={mu}, sigma={sigma}")

# Plot the data and fitted curve
plt.scatter(x_data, y_data, label="Data", color="purple", alpha=0.6)
plt.plot(x_data, gaussian(x_data, A, mu, sigma), label="Fitted Gaussian", color="red")
plt.title("Gaussian Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### **5. Physics and Math Refresher for Curve Fitting**

#### **Linear and Quadratic Equations**
- Linear: $ y = mx + c $
  - $ m $: Slope of the line, representing change in $ y $ per unit $ x $.
  - $ c $: Y-intercept, value of $ y $ when $ x = 0 $.
- Quadratic: $ y = ax^2 + bx + c $
  - $ a $: Controls the curvature of the parabola.
  - $ b $: Controls the tilt of the parabola.
  - $ c $: Y-intercept.

#### **Polynomial Fits**
- Polynomials are sums of terms of the form $ a_nx^n + a_{n-1}x^{n-1} + \ldots + a_0 $.
- Useful for approximating complex data trends.

#### **Gaussian Functions**
- Gaussian curves (bell-shaped) are widely used in astronomy to model spectral lines.
  $$ y = A e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $$
  - $ A $: Amplitude (peak height).
  - $ \mu $: Mean (center of the peak).
  - $ \sigma $: Standard deviation (controls width).

#### **Key Physics Contexts**
- **Gravitational Force**:
  $$ F = G \frac{m_1 m_2}{r^2} $$
  - $ G $: Gravitational constant.
  - $ m_1, m_2 $: Masses of two objects.
  - $ r $: Distance between the objects.

- **Period-Luminosity Relation (Cepheid Variables)**:
  - Longer periods correlate with higher luminosities, aiding in distance measurements.

- **Light and Spectra**:
  $$ E = h\nu = \frac{hc}{\lambda} $$
  - $ E $: Energy.
  - $ \nu $: Frequency.
  - $ \lambda $: Wavelength.

# Explanation of Dataset Columns

1. **`name`**:  
   Unique identifier for the Cepheid variable star, often drawn from astronomical catalogs like OGLE (Optical Gravitational Lensing Experiment). This helps in referencing the specific star across datasets and literature.

2. **`RA0`**:  
   Right Ascension (RA) in decimal degrees. RA is the celestial equivalent of longitude, indicating how far eastward a celestial object is from the vernal equinox, measured along the celestial equator. It helps pinpoint the star's location in the sky.

3. **`Decl0`**:  
   Declination (Decl) in decimal degrees. Declination is the celestial equivalent of latitude, measuring the angular distance of a celestial object north or south of the celestial equator.

4. **`Mode`**:  
   The pulsation mode of the Cepheid star. Common values include:  
   - **`F`**: Fundamental mode, where the entire star pulsates in unison.  
   - **`1`**: First overtone mode, a higher-frequency oscillation.  

5. **`Cloud`**:  
   Indicates the galaxy or stellar cloud where the star resides, for example:  
   - **`LMC`**: Large Magellanic Cloud, a nearby satellite galaxy of the Milky Way.  
   - **`SMC`**: Small Magellanic Cloud, another satellite galaxy of the Milky Way.  

6. **`W`**:  
   Apparent magnitude of the star, a measure of its brightness as seen from Earth. Lower values correspond to brighter stars. This is crucial for understanding the star's luminosity and distance.

7. **`logP1`**:  
   Logarithm (base 10) of the pulsation period in days. The period is the time it takes for the Cepheid to complete one cycle of brightness variation. The logarithmic scale is used for easier interpretation of large ranges of period values.

8. **`VI`**:  
   Color index, calculated as the difference between magnitudes in the V (visible) and I (infrared) bands. This indicates the starâ€™s color, correlating with its temperature:
   - Lower values indicate bluer (hotter) stars.
   - Higher values indicate redder (cooler) stars.

### Analysis of Color Split and Residual Patterns in Cepheid Variables

---

### **1. Wesenheit Index and Dust Extinction**

- **Wesenheit Index (`W`)**: A reddening-free magnitude combining apparent brightness and color to correct for interstellar extinction.
  - Formula:  
    $ W = m_V - R \cdot (m_V - m_I) $

    where:
    - $ m_V $: Visual magnitude.
    - $ m_I $: Infrared magnitude.
    - $ R $: Constant dependent on the extinction law.

- **Overestimated `W`**: 
  - If observed `W` values are systematically lower (blue residuals), extinction corrections are incomplete.
  - Denser regions of interstellar dust/gas absorb and scatter more light, reducing brightness and altering colors.

---

### **2. Differences Between LMC and SMC**

#### **LMC (Large Magellanic Cloud)**

- **Central Region of Mode F**:
  - Concentration of blue residuals suggests higher absorption in the central region.
  - The central part of the LMC contains dense star-forming regions and significant gas and dust.
  - Effects:
    1. **Increased Extinction**: Light from stars is absorbed/scattered by interstellar material.
    2. **Underestimated Extinction Correction**: The expected `W` is higher than observed due to unaccounted absorption.

- **Even Residual Distribution in Mode F and Mode 1**:
  - Indicates relatively uniform ISM absorption across these regions, apart from localized central effects.

#### **SMC (Small Magellanic Cloud)**

- **Shifted Blue Dots to the Left**:
  - Indicates denser gas along the line of sight in the left-region of the SMC.
  - The SMCâ€™s irregular structure leads to pockets of dense gas and dust.
  - Effects:
    1. **Localized Extinction**: Stars in denser regions experience more scattering and absorption.
    2. **Distance-Related Effects**: Fainter appearance of stars due to combined gas density and increased distance.

---

### **3. Split in Colors (Green vs. Purple Points)**

#### **Impact on Apparent Brightness (`W`)**
- Dense interstellar material dims stars more than expected, particularly in the visual band.
- Alters Wesenheit index calculations, shifting stars in magnitude-color space.

#### **Impact on Color Index (`VI`)**
- Dust scatters shorter wavelengths (blue light) more than longer wavelengths (red light).
- Results:
  - Stars behind dense gas appear redder (higher `VI`).
  - LMC Mode F: Central stars dimmer and redder due to higher extinction.
  - SMC left-region stars: Similar effects from localized dense gas.

#### **Spatial Differences in Gas and Dust**
- Variations in gas and dust density within the LMC and SMC lead to:
  - Uneven extinction corrections.
  - Non-uniform distributions of brightness and color indices.

---

### **4. Implications for Distance Measurements**

#### **Cepheid Variables as Standard Candles**
- Accurate distances rely on precise calibration of the period-luminosity relation.
- If extinction is not properly corrected, distances will be systematically underestimated.

#### **Corrections Needed**m
1. Use detailed extinction maps of the LMC and SMC to refine corrections.
2. Model ISM variations to account for gas and dust density differences.
3. Utilize multi-wavelength observations (e.g., ultraviolet, infrared) to better constrain extinction effects.

---


The split in colors and the residual patterns (blue and red dots) align with the hypothesis that light absorption and scattering by the ISM influence the observed properties of Cepheid variables. These effects are localized in specific regions within the LMC and SMC due to variations in gas and dust density, distance, and line-of-sight effects.

Accurate modeling of these phenomena is critical for improving distance measurements to these galaxies and refining the period-luminosity relation for Cepheid variables.
