## 1. Coordinate Conversion (Spherical to Cartesian)
Given a star located in spherical coordinates $r = 5 \, \text{AU}, \, \theta = 60^\circ, \, \phi = 120^\circ$:  
   a. Convert these coordinates into Cartesian (x, y, z).  
   b. Use NumPy to perform the calculations.  

   $$x=r\cdot\sin(\theta)\cdot\cos(\phi)$$
   $$y=r\cdot\sin(\theta)\cdot\sin(\phi)$$
   $$z=r\cdot\cos(\theta)$$

   Write a function which accepts 3 arguments, r, theta and phi and returns x,y,z.

````python
import numpy as np
def conv(r,theta,phi):
    """
    function to convert spherical cordinates to cartesian
    arg: r , theta in degree , phi in degree
    """
    x = r*np.sin(np.deg2rad(theta))*np.cos(np.deg2rad(phi))
    y = r*np.sin(np.deg2rad(theta))*np.sin(np.deg2rad(phi))
    z = r*np.cos(np.deg2rad(theta))
    return x,y,z

conv(r=5,theta=60,phi=120)



````

## 2. Coordinate Conversion (Cartsian to Spherical)

A star's position in the sky can be represented using Cartesian coordinates $(x, y, z)$ or spherical coordinates $r$, $\theta$, $\phi$. 

1. Write a Python function that converts Cartesian coordinates \((x, y, z)\) to spherical coordinates \((r, \theta, \phi)\), where:
   - $r$ is the radial distance:  
     $$
     r = \sqrt{x^2 + y^2 + z^2}
     $$
   - $\theta$ is the polar angle (inclination) in radians:  
     $$
     \theta = \arccos\left(\frac{z}{r}\right) \, \text{if } r \neq 0
     $$
   - $\phi$ is the azimuthal angle in radians:  
     $$
     \phi = \arctan(y, x)
     $$ (Use `np.arctan2`)

2. Write a program to take the Cartesian coordinates of a star as input and print its spherical coordinates.


````python
import numpy as np
def cartsph(x,y,z):
    r = np.sqrt(x**2 +y**2 +z**2)
    theta = np.arccos(z/r)
    phi = np.arctan2(y,x)
    return r,theta,phi
a= input("value of x")
b= input("value of y")
c= input("value of z")
x=float(a)
y=float(b)
z=float(c)
# since value is not provided so useing result of above value to crosscheck
(r,t,p) =cartsph(x,y,z)
print (f"value of r : {r} \n value of theta : {t} \n value of phi : {p} ")


````

## 3. Projectile motion

Write a Python function to calculate the range of a projectile. The function should take the following inputs
1. Initial velocity ($u$) in $m/s$
2. Launch angle ($\theta$)
3. Gravitational acceleration ($g$) in $m/s^2$

The function should calculate and return the range of the projectile using the formula:
$$R=\dfrac{u^2\cdot\sin(2\theta)}{g}$$

Using formatted string, print your result in such a way: 
```For initial velocity of 20 m/s and launch angle of 45 degrees, the range of the projectile is: 40.82 meters```

````python
import numpy as np
def pro(u,thet,g):
    """
    function to calculate the range of projectile
    arg: u = initial velocity, thet= theta (in degrees), g = gravitational accn
    """
    R = u**2 *np.sin(2*np.deg2rad(thet))/g
    Rround = np.round(R,2)
    return Rround
Range = pro(20,45,9.8)
print(f"For initial velocity of 20 m/s and launch angle of 45 degrees, the range of projectile is: {Range} meters")

````

## 6. Star temperature

Create a numpy array of temperatures of 5 stars in Kelvin. Say ```[1000,2000,3000,4000,5000,6000,7000]```
Write a function to calculate their corresponding peak wavelength ($\lambda_{peak}$) using Wein's displacement law: 
$$\lambda_{peak}=\dfrac{2.897\times10^{-3}}{T}$$ 
where $T$ is the temperature in Kelvin and $\lambda_{peak}$ is in meters'

Plot temperature vs peak wavelength.

To make the plot smoother use `np.linspace(1000,7000,100)` and consider 100s of different temperatures between 1000 and 7000 and plot the graph.

````python
ar = np.array([1000,2000,3000,4000,5000,6000,70000])
def peak(T):
    lam = (2.897*10-3)/T
    return lam
Lambda = np.zeros_like(ar, dtype=float)
for i,T in enumerate(ar):
    Lambda[i] =peak(T)
fig,ax = plt.subplots(2,1)
ax[0].plot(ar,Lambda)
ar2 = np.linspace(1000,7000,100)
Lambda2 = np.zeros_like(ar2, dtype=float)
for i,T in enumerate(ar2):
    Lambda2[i] =peak(T)

ax[1].plot(ar2,Lambda2)
````
